```python
import scanpy as sc
import anndata as ad
import numpy as np
```

```python
path_h5ad = "ath_snrna_balanced_flower_seedling_rosette.h5ad"
adata = sc.read_h5ad(path_h5ad)
```

```python
batch_info = adata.obs_names.to_list()
batch_info = [i.split("_")[-1] for i in batch_info]
print(batch_info)
```

```python
print(np.unique(batch_info))
```

```python
adata.obs["batch"] = batch_info
```

```python
# Save
adata.write_h5ad(path_h5ad, compression="gzip")
```
