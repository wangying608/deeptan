# Merge ATAC peaks

```python
import pandas as pd
import scanpy as sc
from deeptan.utils.peaks import MergePeaks, map_peaks_to_ref
```

```python
path_h5ad_control1 = "/mnt/hdd2/homext/wuch/xn2p/data/raw_df/scMultiome/adata_old_control1_ATCK_copy.h5ad"
path_h5ad_control2 = "/mnt/hdd2/homext/wuch/xn2p/data/raw_df/scMultiome/adata_old_control2_CK_copy.h5ad"
```

```python
merge_ = MergePeaks(
    [
        sc.read_h5ad(path_h5ad_control1),
        sc.read_h5ad(path_h5ad_control2),
    ],
    "./",
)
peaks = merge_.merge_peaks()

print(peaks)
print(merge_.check())
```

```python
trans_adata_control1 = map_peaks_to_ref(path_h5ad_control1, peaks)
trans_adata_control2 = map_peaks_to_ref(path_h5ad_control2, peaks)

print(trans_adata_control1.var["feature_types"].value_counts())
print("\n", trans_adata_control2.var["feature_types"].value_counts())
```

```python
print(trans_adata_control1.obs)
print(trans_adata_control1.var)
```

```python
trans_adata_controls = sc.concat([trans_adata_control1, trans_adata_control2], axis=0)
print(trans_adata_controls)
```

```python
print(trans_adata_controls.obs)
print(trans_adata_controls.var)
```

```python
trans_adata_controls_atac = trans_adata_controls[:, : peaks.shape[0]]
new_var = pd.DataFrame(
    {
        "name": trans_adata_controls_atac.var.index.to_list(),
        "feature_types": ["Peaks" for i in range(peaks.shape[0])],
    }
)
new_var.index = trans_adata_controls_atac.var.index
# print(new_var)
trans_adata_controls_atac.var = new_var
```

```python
print(trans_adata_controls_atac)
```

```python
print(trans_adata_controls_atac.var)
```
