```python
ids_df_exp: list[str] = df_exp["obs_names"].unique().to_list()
ids_df_exp = [x.split("_")[-1] for x in ids_df_exp]
print(ids_df_exp)
```

```python
# Check if the ids are in the dict_acc2eco
has_eco = []
ecos = []
for id in ids_df_exp:
    if id in dict_acc2eco.keys():
        has_eco.append(True)
        ecos.append(dict_acc2eco[id])
    else:
        has_eco.append(False)
        # ecos.append(None)
print(has_eco)
print(ecos)
print(len(ecos))
```

```python
# Save df_exp with the eco information
# Pick rows first
df_exp_with_eco = df_exp.filter(has_eco)
# Add eco column to the first column
df_exp_with_eco = pl.DataFrame({"eco": ecos}).hstack(df_exp_with_eco)
print(df_exp_with_eco)
```

```python
df_exp_with_eco.write_parquet(os.path.join(dir_nmic, "exp.matrix_cv0.1_nmicg.has_eco.parquet"))
```

## Random split

```python
import numpy as np
```

```python
df_exp = pl.read_parquet(os.path.join(dir_nmic, "exp.matrix_cv0.1_nmicg.has_ft16.parquet"))
print(df_exp)
```

```python
n_obs = df_exp.shape[0]

# Random split into train, validation and test sets
train_size = int(0.8 * n_obs)
val_size = int(0.1 * n_obs)
test_size = n_obs - train_size - val_size

seeds = [42, 43, 44, 45, 46]
dir_output = os.path.join(dir_bulk, "exp_split")
os.makedirs(dir_output, exist_ok=True)

for seed in seeds:
    np.random.seed(seed)
    shuffled_indices = np.random.permutation(n_obs)
    train_indices = shuffled_indices[:train_size]
    val_indices = shuffled_indices[train_size : (train_size + val_size)]
    test_indices = shuffled_indices[(train_size + val_size) :]
    _df_trn = df_exp[train_indices]
    _df_val = df_exp[val_indices]
    _df_tst = df_exp[test_indices]
    _df_trn.write_parquet(os.path.join(dir_output, f"split_{seed}_0.parquet"))
    _df_val.write_parquet(os.path.join(dir_output, f"split_{seed}_1.parquet"))
    _df_tst.write_parquet(os.path.join(dir_output, f"split_{seed}_2.parquet"))
```
