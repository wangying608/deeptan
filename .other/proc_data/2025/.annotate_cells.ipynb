#### A faster way to search cell types by counting marker genes using Polars

```python
# Convert adata_filtered to a Polars DataFrame.
adata_polars = pl.DataFrame(data=adata_filtered.X.todense())
adata_polars.columns = adata_filtered.var_names
print(adata_polars)
```

```python
# Search for cell types based on the presence of marker genes.
# colname2celltypes = [marker_dict[i] for i in adata_polars.columns]
# print(colname2celltypes)
```

```python
mat_prese = adata_polars.with_columns(
    selected_columns=pl.struct(pl.all()).map_elements(
        lambda row: [col for col, val in row.items() if val[0] > 0],
        return_dtype=pl.List(pl.Utf8),
    )
)
print(mat_prese)
```

```python
print(mat_prese["selected_columns"].head(5))
print(adata_polars.select(mat_prese["selected_columns"][0][0])[0])
```

### Get cell types

```python
barcodes = adata_filtered.obs_names.tolist()

cell_types: dict[str, list[str]] = {}

for i in range(len(barcodes)):
    markers_list = mat_prese["selected_columns"].item(i)
    if len(markers_list) > 0:
        cell_type_list = []
        for marker in markers_list:
            cell_type_list.extend(marker_dict.get(marker))
        cell_types[barcodes[i]] = cell_type_list
    else:
        cell_types[barcodes[i]] = ["Unknown"]
print(cell_types)
```

```python
print(f"Length of dict: {len(cell_types)}")
# Save the cell_types dictionary to a pickle file.
with open("/mnt/bank/sc_sn/GSE235510/sup/cell_types.pkl", "wb") as f:
    pickle.dump(cell_types, f)
```

#### Find the most frequent cell type for each barcode

```python
import numpy as np
```

```python
cts_uniq = {}
cts_unkn = {}
for bc in barcodes:
    uniq_ct = set(cell_types[bc])
    if len(uniq_ct) == 1:
        cts_uniq[bc] = list(uniq_ct)[0]
    else:
        freq_dict = {}
        for ct in cell_types[bc]:
            freq_dict[ct] = freq_dict.get(ct, 0) + 1
        # most_freq_ct = max(freq_dict, key=freq_dict.get)
        # If some cell types has the same count, same the barcode to another dict cts_unkn.
        # Else save it to cts_uniq
        max_counts = max(freq_dict.values())
        most_freq_cts = [ct for ct, count in freq_dict.items() if count == max_counts]
        if len(most_freq_cts) > 1:
            cts_unkn[bc] = most_freq_cts
        else:
            cts_uniq[bc] = most_freq_cts[0]
```

#### Using clustering to determine the most-like cell type for multi-label results

# NEW

```python
import scanpy as sc
import pandas as pd
import numpy as np
import anndata
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
```

```python

# 1. 数据加载与预处理
path_10x_h5 = "/mnt/hdd2/homext/wuch/xn2p/data/raw_df/scMultiome/GSE235510_control.h5"
adata = sc.read_10x_h5(path_10x_h5, gex_only=False)
adata.var_names_make_unique(join="_")

# 2. 质量控制
# 计算基本指标
sc.pp.calculate_qc_metrics(adata, inplace=True)

# 过滤细胞和基因
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)

# 3. 数据标准化
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)

# 4. 特征选择
sc.pp.highly_variable_genes(adata, n_top_genes=2000)
adata = adata[:, adata.var.highly_variable]
```

```python

# 5. QC Metrics
adata.var["mt"] = adata.var_names.str.startswith("MT-")
sc.pp.calculate_qc_metrics(adata, qc_vars=["mt"], inplace=True)

# 6. Regress out covariates
sc.pp.regress_out(adata, ['total_counts', 'pct_counts_mt'])
sc.pp.scale(adata, max_value=10)

# sc.tl.pca(adata, svd_solver='arpack')
sc.pp.neighbors(adata, n_pcs=30, n_neighbors=20)
sc.tl.umap(adata)
sc.tl.leiden(adata, resolution=0.5)
```

```python

# 6. 加载标记基因
marker_dict = pd.read_pickle("/mnt/bank/sc_sn/GSE235510/sup/marker_dict.pkl")

# 7. 差异表达分析
sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon')

# 8. 基于聚类的注释
cell_type_assignments = {}

for cluster in adata.obs['leiden'].cat.categories:
    # 获取该聚类的差异基因
    markers = sc.get.rank_genes_groups_df(adata, group=cluster)
    top_markers = markers.head(20)['names'].tolist()
    
    # 匹配标记基因
    scores = {}
    for cell_type, genes in marker_dict.items():
        overlap = len(set(top_markers) & set(genes))
        scores[cell_type] = overlap
    
    # 选择最佳匹配
    if max(scores.values()) > 0:
        best_match = max(scores, key=scores.get)
        cell_type_assignments[cluster] = best_match
    else:
        cell_type_assignments[cluster] = 'Unknown'

# 9. 添加注释到观察数据
adata.obs['cell_type'] = [
    cell_type_assignments[cluster] 
    for cluster in adata.obs['leiden']
]
```

```python

# 10. 可视化验证
sc.pl.umap(
    adata,
    color=['leiden', 'cell_type'],
    legend_loc='on data',
    frameon=False,
    ncols=1,
    wspace=0.5
)

# 绘制标记基因表达图
selected_genes = list(marker_dict.keys())[:10]
sc.pl.dotplot(
    adata,
    var_names=selected_genes,
    groupby='cell_type',
    dendrogram=True,
    standard_scale='var'
)
```
