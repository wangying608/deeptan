## Read marker gene list

```python
path_marker_genes = "/mnt/bank/sc_sn/GSE235510/sup/advs8042-sup-0002-tables1.xlsx"
```

```python
marker_list1 = pl.read_excel(path_marker_genes, sheet_name="Marker list1")
# Use 1st row as column names
marker_list1.columns = marker_list1.row(0)
# Drop the first row
marker_list1 = marker_list1[1:, :]
marker_list1 = marker_list1.drop("References")

print(marker_list1)
```

```python
marker_list2 = pl.read_excel(path_marker_genes, sheet_name="Marker list 2", has_header=True)
marker_list2 = marker_list2.drop("Reference")

print(marker_list2)
```

```python
marker_list3 = pl.read_excel(path_marker_genes, sheet_name="Marker list3", has_header=True)
marker_list3 = marker_list3.drop("Reference").drop("Software")

print(marker_list3)
```

```python
marker_list_ecd = pl.read_excel(path_marker_genes, sheet_name="Endodermis&cortex development", has_header=True)
marker_list_ecd = marker_list_ecd.drop("Reference")

print(marker_list_ecd)
```

### Concat lists

```python
marker_list = marker_list1.vstack(marker_list2).vstack(marker_list3).vstack(marker_list_ecd)
print(marker_list)
```

### Marker gene dict

```python
marker_dict = {}
for rowx in marker_list.iter_rows():
    if rowx[1] in marker_dict:
        # marker_dict[rowx[1]].append(rowx[2])
        if rowx[2] not in marker_dict[rowx[1]]:
            marker_dict[rowx[1]].append(rowx[2])
    else:
        marker_dict[rowx[1]] = [rowx[2]]
print(marker_dict)
print(len(marker_dict))
```

#### Write marker dict to a pickle file

```python
import pickle

path_marker_dict = "/mnt/bank/sc_sn/GSE235510/sup/marker_dict.pkl"
with open(path_marker_dict, "wb") as f:
    pickle.dump(marker_dict, f)
```
