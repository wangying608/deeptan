## Make data matrices

#### Filter (Threshold of coefficient of variation for removing features [default: 0.1])

```python
import numpy as np

def read_mCXX(file_path: str, colnames: list[str], meth: str = "mCXX"):
    mat_ = pl.read_csv(file_path, separator="\t", has_header=False, null_values="NA", infer_schema_length=10000)
    mat_.columns = colnames
    # Add "mCG_" prefix to the 1st column's data
    prefix_ = meth + "_"
    mat_ = mat_.with_columns((pl.lit(prefix_) + pl.col(colnames[0])).alias(colnames[0]))
    return mat_


def filter_cv(df: pl.DataFrame, threshold: float=0.1, exclude_columns: list[str]=["gene_id"]):
    # Calculate the coefficient of variation (CV)
    np_df_exclude = df.select(pl.exclude(exclude_columns)).to_numpy()
    cv = np.std(np_df_exclude, axis=1) / np.mean(np_df_exclude, axis=1)

    # Filter out features where CV < threshold
    filtered_df = df.filter(cv >= threshold)

    return filtered_df
```

### mCG

```python
path_mCG = "/mnt/bank/1001/gene_body_methylation_data/Araport11_GB_mCG_strict.tsv"
mat_mcg = read_mCXX(path_mCG, colnames, "mCG")
print(mat_mcg.shape)
mat_mcg_f = filter_cv(mat_mcg)
print(mat_mcg_f)
```
