# Pick available barcodes based on given data

```python
print(len(set(adata_ck.obs_names).intersection(set("CK_" + adata_old_control1.obs_names))))
print(len(set(adata_ck.obs_names).intersection(set("CK_" + adata_old_control2.obs_names))))
```

```python
print(len(set(adata_atck.obs_names).intersection(set("AT-CK_" + adata_old_control1.obs_names))))
print(len(set(adata_atck.obs_names).intersection(set("AT-CK_" + adata_old_control2.obs_names))))
```

```python
adata_old_control2_CK = adata_old_control2[("CK_" + adata_old_control2.obs_names).isin(adata_ck.obs_names)].copy()
```

```python
adata_old_control1_ATCK = adata_old_control1[("AT-CK_" + adata_old_control1.obs_names).isin(adata_atck.obs_names)].copy()
```

```python
adata_old_control1_ATCK.obs_names = "AT-CK_" + adata_old_control1_ATCK.obs_names
adata_old_control2_CK.obs_names = "CK_" + adata_old_control2_CK.obs_names
```

```python
print(adata_old_control1_ATCK)
print(adata_old_control1_ATCK.obs.head())
print(adata_old_control2_CK)
print(adata_old_control2_CK.obs.head())
```

```python
if set(adata_old_control1_ATCK.obs_names) != set(adata_atck.obs_names):
    print("两个 AnnData 对象的条形码集合不相同！")
else:
    print("两个 AnnData 对象的条形码集合相同！")
```

```python
if set(adata_old_control2_CK.obs_names) != set(adata_ck.obs_names):
    print("两个 AnnData 对象的条形码集合不相同！")
else:
    print("两个 AnnData 对象的条形码集合相同！")
```

```python
adata_old_control1_ATCK_copy = adata_old_control1_ATCK.copy()
adata_old_control2_CK_copy = adata_old_control2_CK.copy()
```

```python
print(adata_atck.is_view)  # Returns True if it's a view
```

```python
for obs_key in adata_atck.obs_keys():
    if obs_key not in adata_old_control1_ATCK_copy.obs_keys():
        adata_old_control1_ATCK_copy.obs[obs_key] = adata_atck.obs[obs_key]
```

```python
for obs_key in adata_ck.obs_keys():
    if obs_key not in adata_old_control2_CK_copy.obs_keys():
        adata_old_control2_CK_copy.obs[obs_key] = adata_ck.obs[obs_key]
```

```python
print(adata_old_control1_ATCK_copy, "\n")
print(adata_old_control1_ATCK_copy.obs.head())
```

```python
print(adata_old_control2_CK_copy, "\n")
print(adata_old_control2_CK_copy.obs.head())
```
